# Lab 3 Altair Visualization

For this lab I used the Seattle Library dataset to make two different Altair charts. I wanted to look at a couple parts of the data and practice building interactive visualizations.

```python
import pandas as pd
import altair as alt
import requests
import io
```

```python
url = "https://raw.githubusercontent.com/manika-lamba/SP26-LIS4_5693/refs/heads/main/lab-assignments/lab-3/Seattle-Library_2015-2021.csv"

seattle_df = pd.read_csv(url)

seattle_df.columns
```

## Loading the dataset

I loaded the Seattle Library dataset into a pandas dataframe called `seattle_df`. After that I checked the columns so I could choose two variables to visualize.


```python
year_counts = seattle_df.groupby("CheckoutYear")["Checkouts"].sum().reset_index()

year_counts.head()
```

```python
line_chart = alt.Chart(year_counts).mark_line(point=True).encode(
    x=alt.X("CheckoutYear:O", title="Checkout Year"),
    y=alt.Y("Checkouts:Q", title="Total Checkouts"),
    tooltip=["CheckoutYear", "Checkouts"]
).properties(
    title="Total Library Checkouts by Year"
).interactive()

line_chart
```

## First visualization

For my first chart I looked at total checkouts by year. I grouped the data by `CheckoutYear` and summed the `Checkouts` column so I could see how library activity changed over time.

```python
material_counts = seattle_df.groupby("MaterialType")["Checkouts"].sum().reset_index()

material_counts = material_counts.sort_values("Checkouts", ascending=False).head(10)

material_counts
```

```python
point_chart = alt.Chart(material_counts).mark_point(size=120).encode(
    x=alt.X("MaterialType:N", sort="-y", title="Material Type"),
    y=alt.Y("Checkouts:Q", title="Total Checkouts"),
    color="MaterialType:N",
    tooltip=["MaterialType", "Checkouts"]
).properties(
    title="Top Material Types by Checkouts"
).interactive()

point_chart
```

## Second visualization

For the second chart I grouped the data by `MaterialType` and summed total checkouts. I kept the top material types so the chart would be easier to read and compare.

```python
combined_chart = line_chart | point_chart
combined_chart
```

## Combined visualization

After creating both charts, I placed them side by side using horizontal concatenation. This makes it easier to compare the trends in checkouts over time with the distribution of material types.
